## Task framing

We solve a binary classification problem.

Goal: predict whether a customer is likely to churn.

Target column: `Churn`.

Target values:
- `No`: customer did not churn.
- `Yes`: customer churned.

Positive class: `Yes`.

Business interpretation:
- False Positive: model predicts churn, but the customer would not actually churn.
  - Possible cost: unnecessary retention offer, discount, or sales effort.
- False Negative: model predicts no churn, but the customer actually churns.
  - Possible cost: lost customer and lost future revenue.

For churn prediction, recall for the positive class is important because the business usually wants to detect as many at-risk customers as possible.

Precision is also important because too many false positives can waste retention budget.

Accuracy alone is not enough because the dataset is imbalanced:
- majority class `No` is about 73.46%;
- a naive model predicting only `No` would already get about 73.46% accuracy.

Initial main metrics:
- Recall for `Churn = Yes`
- Precision for `Churn = Yes`
- F1-score for `Churn = Yes`
- ROC-AUC as a threshold-independent ranking metric

Later we may also look at:
- confusion matrix
- PR-AUC
- threshold tuning on validation data, not on test data
